```python
import math

def binomial_coefficient(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

def Formula(n):
    if n == 0:
        return "1"
    elif n == 1:
        return "a+b"
    elif n < 0:
        return "1/(" + Formula(-n) + ")"
    else:
        terms = []
        for k in range(n+1):
            a_term = "" if n-k == 0 else ("a" if n-k == 1 else "a^" + str(n-k))
            b_term = "" if k == 0 else ("b" if k == 1 else "b^" + str(k))
            if k == 0 or k == n:
                terms.append(a_term + b_term)
            else:
                terms.append(str(binomial_coefficient(n, k)) + a_term + b_term)
        return "+".join(terms)
```
